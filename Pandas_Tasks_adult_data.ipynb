Create a new column 'income_binary' where <=50K is 0 and >50K is 1.

```python
df['income_binary'] = df['income'].apply(lambda x: 0 if x == '<=50K' else 1)
display(df[['income', 'income_binary']].head())
```

```python
bins = [0, 30, 60, df['age'].max()]
labels = ['Youth', 'Adult', 'Senior']
df['age_group'] = pd.cut(df['age'], bins=bins, labels=labels, right=False)
display(df[['age', 'age_group']].head())
```

Create a new column 'age_group' categorizing ages into 'Youth' (< 30), 'Adult' (30-60), and 'Senior' (60+).

```python
bins = [0, 30, 60, df['age'].max()]
labels = ['Youth', 'Adult', 'Senior']
df['age_group'] = pd.cut(df['age'], bins=bins, labels=labels, right=False)
display(df[['age', 'age_group']].head())
```

Rename the 'educational-num' column to 'education_num'.

```python
df = df.rename(columns={'educational-num': 'education_num'})
display(df.head())
```

Replace the missing values represented by '?' in 'workclass' and 'occupation' columns with 'Unknown'.

```python
df['workclass'] = df['workclass'].replace('?', 'Unknown')
df['occupation'] = df['occupation'].replace('?', 'Unknown')
display(df.head())
```

Calculate the average age for each workclass.

```python
average_age_by_workclass = df.groupby('workclass')['age'].mean()
display(average_age_by_workclass)
```

```python
total_capital_gain_by_gender = df.groupby('gender')['capital-gain'].sum()
display(total_capital_gain_by_gender)
```

**Chart 1: Income Distribution by Education Level**
Generate a stacked bar chart to visualize the distribution of income categories (<=50K, >50K) across different education levels. Include observations about how education might influence income.

```python
import matplotlib.pyplot as plt
import seaborn as sns
print("Libraries matplotlib.pyplot and seaborn imported.")
```

```python
crosstab_df = pd.crosstab(df['education'], df['income'])
crosstab_df.plot(kind='bar', stacked=True, figsize=(12, 7))
plt.title('Income Distribution by Education Level')
plt.xlabel('Education Level')
plt.ylabel('Number of Individuals')
plt.xticks(rotation=45, ha='right')
plt.tight_layout()
plt.show()
```

 Chart 2: Average Hours per Week by Gender and Income

Generate a grouped bar chart to compare the average hours worked per week between males and females, separated by income level (<=50K, >50K).

```python
average_hours_by_gender_income = df.groupby(['gender', 'income'])['hours-per-week'].mean().unstack()
display(average_hours_by_gender_income)
```

```python
average_hours_by_gender_income.plot(kind='bar', figsize=(10, 6), rot=0)
plt.title('Average Hours per Week by Gender and Income')
plt.xlabel('Gender')
plt.ylabel('Average Hours per Week')
plt.legend(title='Income')
plt.tight_layout()
plt.show()
```

Chart 4: Capital Gain vs. Hours per Week


Create a scatter plot comparing 'capital-gain' and 'hours-per-week' to see if there's any visible correlation or patterns.


```python
plt.figure(figsize=(12, 8))
sns.scatterplot(x='hours-per-week', y='capital-gain', data=df, alpha=0.6)
plt.title('Capital Gain vs. Hours per Week')
plt.xlabel('Hours per Week')
plt.ylabel('Capital Gain')
plt.grid(True, linestyle='--', alpha=0.7)
plt.show()

```

```python
plt.figure(figsize=(14, 8))
sns.boxplot(x='education_num', y='hours-per-week', hue='income', data=df)
plt.title('Hours per Week Distribution by Education Level and Income')
plt.xlabel('Education Number')
plt.ylabel('Hours per Week')
plt.xticks(rotation=45, ha='right')
plt.tight_layout()
plt.show()
```

Chart 5: Marital Status vs. Income

Generate a stacked bar chart showing the breakdown of income categories (<=50K, >50K) within each 'marital-status'.


```python
marital_income_crosstab = pd.crosstab(df['marital-status'], df['income'])
display(marital_income_crosstab)
```

```python
marital_income_crosstab.plot(kind='bar', stacked=True, figsize=(12, 7))
plt.title('Income Distribution by Marital Status')
plt.xlabel('Marital Status')
plt.ylabel('Number of Individuals')
plt.xticks(rotation=45, ha='right')
plt.tight_layout()
plt.show()

```

The key insights and observations drawn from the generated charts are:

1.  **Income Distribution by Education Level**: There is a strong positive correlation between higher education levels and the likelihood of earning over \$50K. Individuals with advanced degrees (e.g., Prof-school, Doctorate, Masters, Bachelors) show a significantly higher proportion of high-income earners, whereas individuals with education levels up to HS-grad predominantly earn \$50K or less.
2.  **Average Hours per Week by Gender and Income**: Males, regardless of income level, tend to work more hours per week than females. For both genders, earning over \$50K is associated with working more hours per week. This gender gap in average hours worked is more noticeable in the >\$50K income bracket.
3.  **Age Distribution**: The dataset's age distribution is right-skewed, indicating a higher concentration of younger individuals. The most common age range appears to be between 20 and 40 years.
4.  **Workclass Proportions**: The 'Private' workclass is dominant, comprising a large majority of the dataset (around 69.4%). Other significant workclasses include 'Self-emp-not-inc', 'Local-gov', and 'Unknown', though they are considerably smaller than 'Private'.
5.  **Income Distribution by Marital Status**: 'Married-civ-spouse' individuals show the highest number of earners in both income categories, with a substantial portion earning over \$50K. Conversely, 'Never-married' individuals overwhelmingly earn \$50K or less, with a very small proportion earning over \$50K. Most other marital statuses (e.g., Divorced, Separated, Widowed, Married-spouse-absent) also predominantly fall into the \$50K or less income bracket.
